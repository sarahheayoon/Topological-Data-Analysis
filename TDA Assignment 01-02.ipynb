```python
import matplotlib.pyplot as plt
# ploting random (x,y) 
plt.plot([1,2,3,4],[1,5,1,0])
plt.xlabel('integers') #label
plt.show()
```

```python
import numpy as np
# evenly sampled time at 0.2 intervals
t = np.arange(0, 5, 0.2) #from 0 to 5 with 0.2 intervals

# given function y = sin(t)
y = np.sin(t)

# plot t versus sin(t)
plt.plot(t, y, 'bo') #'bo' is for the blue dots
plt.plot(t, y) #connected with lines
plt.show()
```

# Assignment 02: plot (sin(kh), sin((k+1)h)) for k = 0, 1, ..., 50 
- values of h = 0.01, 0.1, 1, 10, 0.57, pi, pi+0.1, 22/7

```python
# evenly sampled time at h interval where h = 0.01
h = 0.01
k = np.arange(0, 100, 1)
x = np.sin(k*h)
y = np.sin((k+1)*h)

# equal aspect ratio
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_aspect('equal')

# plot t versus sin(t)
plt.plot(x, y, 'bo')
#plt.plot(x, y)
plt.show()
```
